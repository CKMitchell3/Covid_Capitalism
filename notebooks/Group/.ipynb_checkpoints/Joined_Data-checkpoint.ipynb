```python
#Import all of the necessary libraries

import os
import pandas as pd
import alpaca_trade_api as tradeapi

from dotenv import load_dotenv

%matplotlib inline
```

```python
pandemic_portfolio = pd.read_csv("../../data/rawdata/Pandemic_Stocks_Raw_Data.csv", index_col = "time", parse_dates = True, infer_datetime_format = True)

pandemic_portfolio.head(10)
```

```python
tsla_gme_portfolio = pd.read_csv("../../data/rawdata/tesla_game_joined.csv", index_col = "time", parse_dates = True, infer_datetime_format = True)

tsla_gme_portfolio.head(10)
```

```python
index_closing_prices = pd.read_csv("../../data/rawdata/Index_Closing_Prices2.csv", index_col = "Date", parse_dates = True, infer_datetime_format = True)

index_closing_prices.head(10)
```

```python
bluechip_portfolio = pd.read_csv("../../data/rawdata/Bluechips_Closing_Prices.csv", index_col = "time", parse_dates = True, infer_datetime_format = True)

bluechip_portfolio.head(10)
```

```python
crypto_portfolio = pd.read_csv("../../data/rawdata/Crypto_Closing_Prices2.csv", index_col = "Date", parse_dates = True, infer_datetime_format = True)

crypto_portfolio.head(10)
```

```python
total_portfolio = pd.concat([pandemic_portfolio, tsla_gme_portfolio, bluechip_portfolio], axis = 'columns', join = 'inner')

total_portfolio.head(10)
```

```python
csv_portfolio = pd.concat([index_closing_prices, crypto_portfolio], axis = 'columns', join = 'inner')

csv_portfolio.head(10)
```

```python
csv_portfolio.reset_index(inplace=True)
```

```python
total_portfolio.reset_index(inplace=True)
```

```python
final_portfolio = pd.concat([total_portfolio, csv_portfolio], axis = 'columns', join = 'inner')

final_portfolio.head(10)
```

```python
final_portfolio.drop(columns = "time", inplace=True)
```

```python
final_portfolio.head(10)
```

```python
final_portfolio.set_index("Date", inplace=True)
```

```python
final_portfolio.head(10)
```

```python
final_portfolio = final_portfolio[['SP500', 'Nasdaq', 'BA', 'DAL', 'MGM', 'WYNN', 'Tesla', 'Gamestop', 'AAPL', 'AMZN', 'BRK.A', 'MSFT', 'BTC', 'ETH']]
```

```python
final_portfolio.head()
```

```python
final_portfolio.dropna(inplace=True)
```

```python
final_portfolio.to_csv("Final_Portfolio.csv")
```
